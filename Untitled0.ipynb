```python
#Residual plot to check for linearity
from sklearn.linear_model import LinearRegression
#run regression
def residual_plot(country_UR, country):
  X = np.array(country_UR.index).reshape((-1,1))
  y = np.array(country_UR.Value)
  reg = LinearRegression().fit(X, y)
  #Generated Predictions
  y_predicted = reg.predict(X)

  #Plot Our Actual and Predicted Values
  fig,ax = plt.subplots(figsize =(9,7))
  residuals = y-y_predicted
  plt.plot(X,residuals, 'o', color='darkblue')
  plt.title(f"Residual Plot of {country}")
  plt.xlabel("Monthly Observation")
  plt.ylabel("Residual")
```

```python
def ARIMA_MLE(df, column, order):
  model = ARIMA(df[column], order = order)
  model_fit = model.fit(disp=0, method = 'mle')
  print(model_fit.summary())
  plt.rcParams.update({'figure.figsize':(10,5), 'figure.dpi':90})
  residuals = pd.DataFrame(model_fit.resid)
  fig, ax = plt.subplots(1,2)
  residuals.plot(title="Residuals", ax=ax[0])
  residuals.plot(kind='kde', title='Density', ax=ax[1])
  return plt.show()

def ARIMA_D(df, column, order):
  model = ARIMA(df[column], order = order)
  model_fit = model.fit(disp=0)
  return print(model_fit.summary())
```

```python
def get_stationarity(timeseries):
    
    # rolling statistics
    rolling_mean = timeseries.rolling(window=12).mean()
    rolling_std = timeseries.rolling(window=12).std()
    
    # rolling statistics plot
    original = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(rolling_mean, color='red', label='Rolling Mean')
    std = plt.plot(rolling_std, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    # Dickey–Fuller test:
    result = adfuller(timeseries['Value'])
    print('ADF Statistic: {}'.format(result[0]))
    print('p-value: {}'.format(result[1]))
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t{}: {}'.format(key, value))
```
